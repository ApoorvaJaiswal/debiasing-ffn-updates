```python
import json 
f = open("stereoset-data.json")
stereoset_data = json.load(f)
```

```python
intersentence_race = list(filter(lambda x: x["bias_type"] == "race",stereoset_data["data"]['intersentence']))
intersentence_gender = list(filter(lambda x: x["bias_type"] == "gender",stereoset_data["data"]['intersentence']))
intersentence_religion = list(filter(lambda x: x["bias_type"] == "religion",stereoset_data["data"]['intersentence']))
intersentence_profession = list(filter(lambda x: x["bias_type"] == "profession",stereoset_data["data"]['intersentence']))
```

```python
def get_antistereotype_sentences(intersentence_dict):
    context = intersentence_dict["context"]
    anti_stereotype_sentences = []
    for sentence in intersentence_dict["sentences"]:
        if sentence["gold_label"] == "anti-stereotype":
            return context + " " + sentence["sentence"]

print(len(list(map(get_antistereotype_sentences, intersentence_race))))
print(len(list(map(get_antistereotype_sentences, intersentence_gender))))
print(len(list(map(get_antistereotype_sentences, intersentence_religion))))
print(len(list(map(get_antistereotype_sentences, intersentence_profession))))

antistereotype_sentences = list(map(get_antistereotype_sentences, stereoset_data["data"]['intersentence']))

with open('antistereotype_sentences.txt', 'w') as f:
    for line in antistereotype_sentences:
        f.write(f"{line}\n")
```
